```python
#importing the models

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
RF=RandomForestClassifier()
RF.fit(x_train,y_train)
y_predict=RF.predict(x_test)

#Prediction summary
print(classification_report(y_test,y_predict))

#Accuracy score
RF_SC=accuracy_score(y_predict,y_test)
print(f"{round(RF_SC*100,2)}% accuracy")
```

```python
LR=LogisticRegression()
LR.fit(x_train,y_train)
y_predict=LR.predict(x_test)

#Prediction summary
print(classification_report(y_test,y_predict))

#Accuracy score
LR_SC=accuracy_score(y_predict,y_test)
print(f"{round(LR_SC*100,2)}% accuracy")
```

```python
KNN=KNeighborsClassifier(n_neighbors=40,p=2,metric='euclidean')
KNN.fit(x_train,y_train)
y_predict=KNN.predict(x_test)

print(classification_report(y_test,y_predict))

KNN_SC=accuracy_score(y_predict,y_test)
print(f"{round(KNN_SC*100,2)}% accuracy")
```

```python
score=[KNN_SC,RF_SC,LR_SC]
models=pd.DataFrame({
    'Algorithm Model':['K Nearest Neighbour','Random Forest','Logistic Regression'],'Score':score})
models.sort_values(by='Score',ascending=False)
```
